## Обучение и проверка модели

### Создание функции, формирующей обученные модели

```python
# Создание функции regression_select(), обучающей модели для
# задач регрессии с перебором параметров, кросс-валидацией
# и выводом метрик качества
# 1. Объявление функции regression_select() и задание
# параметров данной функции
def regression_select(
    model_name,
    features_train_income,
    target_train_income,
    depth_treshold=None,
    depth_step=None,
    leaf_treshold=None,
    leaf_step=None,
    estimators_treshold=None,
    estimators_step=None,
    scoring_income=None,
    cv_income=5,
    best_cvs=3000,
    rs=777
):
    # A.1. Логическое условие для отбора и обучения
    # модели решающего дерева
    if model_name == DecisionTreeRegressor:
        # A.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями максимальной глубины дерева,
        # минимального количества выборок в листьях, количества блоков
        # кросс-валидации, метрик качества
        data_value = []
        # A.2.2. Создание переменной со списком названий колонок
        columns_name = [
            'depth_value',
            'leaf_value',
            'number_cv_blocks',
            'rmse'
        ]
        # A.3. Цикл перебора значений максимальной
        # глубины дерева
        for depth in range(5, depth_treshold, depth_step):
            # A.4. Цикл перебора значений минимального
            # количества выборок в листьях
            for leaf in range(5, leaf_treshold, leaf_step):
                # A.5.1. Создание объекта структуры данных
                model = model_name(
                    max_depth=depth,
                    min_samples_leaf=leaf,
                    random_state=rs
                )
                # A.5.2. Запуск кросс-валидации и формирование
                # массива оценок выбранной метрики качества
                cvs = cross_val_score(
                    model,
                    features_train_income,
                    target_train_income,
                    cv=cv_income,
                    scoring=scoring_income
                )
                # A.5.3. Вычисление среднего значения
                # полученной метрики качества
                mean_cvs = np.abs(cvs.mean())
                # A.5.4. Запуск обучения модели на обучающей выборке
                model.fit(features_train_income, target_train_income)
                # A.5.5. Заполнение списка в переменной data_value
                # значениями максимальной глубины дерева, минимального
                # количества выборок в листьях, количества блоков
                # кросс-валидации, метрик качества
                data_value.append(
                    [
                        depth,
                        leaf,
                        cv_income,
                        mean_cvs.round(decimals=2)
                    ]
                )
                # A.6.1. Логическое условие сохранения параметров модели
                # с лучшим значением mean_cvs
                if mean_cvs < best_cvs:
                    # A.6.2. Сохранение текущего значения mean_cvs
                    best_cvs = mean_cvs
                    # A.6.3. Сохранение лучшей обученной модели
                    best_model = model
        # A.7.1. Формирование датафрейма с комбинациями максимальной глубины
        # дерева, минимального количества выборок в листьях, количества блоков
        # кросс-валидации, полученными значениями метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # A.7.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['rmse'] == new_data['rmse'].min()
        ].reset_index(drop=True)
        # A.8. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение квадратного корня из среднего квадратичного отклонения (RMSE): ', df_cvs['rmse'][0], ' при значениях параметров:', '\n',
            '- максимальная глубина дерева: ', df_cvs['depth_value'][0], '\n',
            '- минимальное количество выборок в листе: ', df_cvs['leaf_value'][0], '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # A.9. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        # A.10. Возвращение функции
        return best_model
# --------------------------------------------------------------------------------------------------------
    # B.1. Логическое условие для отбора и обучения
    # модели случайного леса
    if model_name == RandomForestRegressor:
        # B.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями количества деревьев в лесу, 
        # максимальной глубины дерева, минимального количества
        # выборок в листьях, метрик качества
        data_value = []
        # B.2.2. Создание переменной со списком названий колонок
        columns_name = [
            'number_tree',
            'depth_value',
            'leaf_value',
            'number_cv_blocks',
            'rmse'
        ]
        # B.3. Цикл перебора значений максимальной глубины дерева
        for estim in range(2, estimators_treshold, estimators_step):
            # B.4. Цикл перебора значений максимальной глубины дерева
            for depth in range(5, depth_treshold, depth_step):
                # B.5. Цикл перебора значений минимального
                # количества выборок в листьях
                for leaf in range(5, leaf_treshold, leaf_step):
                    # B.6.1. Создание объекта структуры данных
                    model = model_name(
                        n_estimators=estim,
                        max_depth=depth,
                        min_samples_leaf=leaf,
                        random_state=rs
                    )
                    # B.6.2. Запуск кросс-валидации и формирование
                    # массива оценок выбранной метрики качества
                    cvs = cross_val_score(
                        model,
                        features_train_income,
                        target_train_income,
                        cv=cv_income,
                        scoring=scoring_income
                    )
                    # B.6.3. Вычисление среднего значения
                    # полученной метрики качества
                    mean_cvs = np.abs(cvs.mean())
                    # B.6.4. Запуск обучения модели на обучающей выборке
                    model.fit(features_train_income, target_train_income)
                    # B.6.5. Заполнение списка в переменной data_value
                    # значениями количества деревьев в лесу, максимальной
                    # глубины дерева, минимального количества выборок в
                    # листьях, количества блоков кросс-валидации, метрик
                    # качества
                    data_value.append(
                        [
                            estim,
                            depth,
                            leaf,
                            cv_income,
                            mean_cvs.round(decimals=2)
                        ]
                    )
                    # B.7.1. Логическое условие сохранения параметров модели
                    # с лучшим значением mean_cvs
                    if mean_cvs < best_cvs:
                        # B.7.2. Сохранение текущего значения mean_cvs
                        best_cvs = mean_cvs
                        # B.7.3. Сохранение лучшей обученной модели
                        best_model = model
        # B.8.1. Формирование датафрейма с комбинациями количества деревьев
        # в лесу, максимальной глубины дерева, минимального количества выборок
        # в листьях, количества блоков кросс-валидации, полученными значениями
        # метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # B.8.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['rmse'] == new_data['rmse'].min()
        ].reset_index(drop=True)
        # B.9. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение квадратного корня из среднего квадратичного отклонения (RMSE): ', df_cvs['rmse'][0], ' при значениях параметров:', '\n',
            '- количество деревьев в лесу: ', df_cvs['number_tree'][0], '\n',
            '- максимальная глубина дерева: ', df_cvs['depth_value'][0], '\n',
            '- минимальное количество выборок в листе: ', df_cvs['leaf_value'][0], '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # B.10. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        # B.11. Возвращение функции
        return best_model
# --------------------------------------------------------------------------------------------------------
    # C.1. Логическое условие для отбора и обучения
    # модели линейной регрессии
    if model_name == LinearRegression:
        # C.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями количества блоков кросс-валидации,
        # метрик качества
        data_value = []
        # C.2.2. Создание переменной со списком названий колонок
        columns_name = [
            'number_cv_blocks',
            'rmse'
        ]
        # C.2.3. Создание объекта структуры данных
        model = model_name()
        # C.2.4. Запуск кросс-валидации и формирование
        # массива оценок выбранной метрики качества
        cvs = cross_val_score(
            model,
            features_train_income,
            target_train_income,
            cv=cv_income,
            scoring=scoring_income
        )
        # C.2.5. Вычисление среднего значения
        # полученной метрики качества
        mean_cvs = np.abs(cvs.mean())
        # C.2.6. Запуск обучения модели на обучающей выборке
        model.fit(features_train_income, target_train_income)
        # C.2.7. Заполнение списка в переменной data_value
        # значениями количества блоков кросс-валидации,
        # метрик качества
        data_value.append(
            [
                cv_income,
                mean_cvs.round(decimals=2)
            ]
        )
        # C.3.1. Логическое условие сохранения параметров модели
        # с лучшим значением mean_cvs
        if mean_cvs < best_cvs:
            # C.3.2. Сохранение текущего значения mean_cvs
            best_cvs = mean_cvs
            # C.3.3. Сохранение лучшей обученной модели
            best_model = model
        # C.4.1. Формирование датафрейма с комбинациями количества блоков
        # кросс-валидации, полученными значениями метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # C.4.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['rmse'] == new_data['rmse'].min()
        ].reset_index(drop=True)
        # C.5. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение квадратного корня из среднего квадратичного отклонения (RMSE): ', df_cvs['rmse'][0], ' при значениях параметров:', '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # C.6. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        # C.7. Возвращение функции
        return best_model
```
